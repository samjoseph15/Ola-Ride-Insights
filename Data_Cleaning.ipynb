## DataCleaning and PreProcessing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("OLA_DataSet.xlsx")
df
```

```python
print(df.duplicated().sum()) # There is no duplicated values
print(df.shape)
df.isnull().sum().sort_values(ascending=False)
```

```python
df1 = df.drop(["Incomplete_Rides_Reason","Vehicle Images"],axis=1,inplace=False)

df1["Date"] = pd.to_datetime(df1["Date"], errors="coerce").dt.date
df1['Booking_ID'] = df1['Booking_ID'].str.extract('(\d+)') # will remove Variables
df1['Customer_ID'] = df1['Customer_ID'].str.extract('(\d+)')

df1
```

```python
ride_volume = df1.groupby(df1["Date"]).size()
print(ride_volume.head())
```

```python
df1.isnull().sum().sort_values(ascending=False)
```

```python
df1.describe()
```

```python
df1["Driver_Ratings"] = df1["Driver_Ratings"].fillna(df1["Driver_Ratings"].mean().round(1))
df1["Customer_Rating"] = df1["Customer_Rating"].fillna(df1["Customer_Rating"].mean().round(1))
df1["V_TAT"] = df1["V_TAT"].fillna(df1["V_TAT"].median())
df1["C_TAT"] = df1["C_TAT"].fillna(df1["C_TAT"].median())
df1["Ride_Distance"] = df1["Ride_Distance"].fillna(df1["Ride_Distance"].median())
df1["Booking_Value"] = df1["Booking_Value"].fillna(df1["Booking_Value"].median())
df1["Incomplete_Rides"] = df1["Incomplete_Rides"].map({"Yes": 1, "No": 0}).fillna(0)
df1["Payment_Method"] = df1["Payment_Method"].fillna("Unknown")
df1["Canceled_Rides_by_Customer"] = df1["Canceled_Rides_by_Customer"].fillna("Unknown")
df1["Canceled_Rides_by_Driver"] = df1["Canceled_Rides_by_Driver"].fillna("Unknown")
```

### Encoding

```python
print(df1["Pickup_Location"].unique()) # Nominal Data - one hot encoding
print("---------------------------------")
print(df1["Drop_Location"].unique()) # Nominal Data - one hot encoding
print("---------------------------------")
```

```python
print(df1["Vehicle_Type"].unique()) # Nominal Data - one hot encoding
print("---------------------------------")
print(df1["Booking_Status"].unique()) # nominal encoding
print("---------------------------------")
print(df1["Canceled_Rides_by_Customer"].unique()) # Nominal Data - one hot encoding
print("---------------------------------")
print(df1["Canceled_Rides_by_Driver"].unique()) # Nominal Data 
print("---------------------------------")
print(df1["Payment_Method"].unique()) # Nominal data
```

```python
df_encoded = pd.get_dummies(df1, columns=[
                                        "Pickup_Location",
                                        "Drop_Location",
                                        "Vehicle_Type",
                                        "Booking_Status", 
                                        "Canceled_Rides_by_Customer",
                                        "Canceled_Rides_by_Driver",
                                        "Payment_Method"], dtype="int")
df_encoded
```
